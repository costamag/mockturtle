```python
import numpy as np
N = [k for k in range(10,18,1)]
S = [0.152000, 0.267000, 0.520000, 0.991000, 1.925000, 3.784000, 6.909000, 12.594999] 
P = [0.004000, 0.005000, 0.007000, 0.014000, 0.024000, 0.042000, 0.081000, 0.148000]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(N, 100*abs(np.array(S)-np.array(P))/S, "-o")
plt.xticks(N, [2**k for k in N])
plt.title("Speedup", fontsize=25)
plt.ylabel(r"$\frac{T_S-T_P}{T_S}[\%]$", fontsize=20)
plt.xlabel("Number of simulation patterns", fontsize=20)
```

```python
plt.plot(N, np.array(S), "-o")
plt.xticks(N, [2**k for k in N])
plt.title("Simulation Time ", fontsize=25)
plt.ylabel(r"$T_S[ms]$", fontsize=20)
plt.xlabel("Number of simulation patterns", fontsize=20)
```

```python
plt.plot(N, np.array(P), "-o", label="AIG")
plt.xticks(N, [2**k for k in N])
plt.title("Simulation Time ", fontsize=25)

plt.plot(N, np.array(S), "-o",label="(POP,XNOR)" )
plt.xticks(N, [2**k for k in N])
plt.ylabel(r"$T_S[ms]$", fontsize=20)
plt.xlabel("Number of simulation patterns", fontsize=20)
plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(111)
x = N
ax.semilogy(N, np.array(P), '-o', label="AIG", linewidth=4, markersize=13)
ax.semilogy(N, np.array(S), '-^', label="POPCOUNT", linewidth=4, markersize=13)


ax.set_title("Simulation Time vs Simulation Size [ms]", fontsize=20)
ax.set_xlabel(r"$\log_2(N)$", fontsize=18)
ax.set_ylabel("Time [ms]", fontsize=18)
ax.set_xticks(N, N)

ax.grid(True)
ax.legend(loc=0, fontsize=12)
fig.savefig('Basic.png', dpi=300)
```
