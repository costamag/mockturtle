```python
import numpy as np
N = [k for k in range(10,18,1)]
S = [0.152000, 0.267000, 0.520000, 0.991000, 1.925000, 3.784000, 6.909000, 12.594999] 
P = [0.004000, 0.005000, 0.007000, 0.014000, 0.024000, 0.042000, 0.081000, 0.148000]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(N, abs(np.array(S)-np.array(P))/S, "-o")
plt.xticks(N, N)
```

```python
plt.plot(N, np.array(S), "-o")
plt.xticks(N, N)
```

```python
plt.plot(N, np.array(P), "-o")
plt.xticks(N, N)
plt.yticks(P, P)
```
